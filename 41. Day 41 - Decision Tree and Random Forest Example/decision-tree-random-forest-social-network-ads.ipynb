# **Social Network Ads - Decision Tree and Random Forest**

# **1. Decision Tree**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

> **1. Data Pre-Processing :**

```python
df = pd.read_csv('/kaggle/input/social-network-ads/Social_Network_Ads.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
X = df.drop('Purchased', axis=1)
Y = df['Purchased']
```

```python
from  sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

> **2. Fitting Decision Tree Algorithm using Training Data :**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion='entropy')
model = classifier.fit(x_train, y_train)
```

> **3. Predicting the Test Set Results :**

```python
y_pred = classifier.predict(x_test)
```

> **4. Evaluate Performance of the Model :**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
mat = confusion_matrix(y_test, y_pred)
mat
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
```

> **5. Plot the Decision Tree :**

```python
from sklearn import tree
```

```python
plt.figure(figsize = (10, 15))
tree.plot_tree(model, filled=True)
```

# **2. Random Forest**

> **1. Fitting Random Forest Classifier on Training Data :**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 600)
rfc.fit(x_train, y_train)
```

> **2. Predicting the Test Set Results :**

```python
y_pred = rfc.predict(x_test)
```

> **3. Model Evaluation :**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
mat = confusion_matrix(y_test, y_pred)
mat
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
```

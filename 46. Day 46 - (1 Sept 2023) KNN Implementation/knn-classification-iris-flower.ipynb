# **IRIS Flower Prediction using KNN Classification**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

> **1. Data Pre-processing :**

```python
df = pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
```

```python
df
```

```python
df.info()
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
df['species'] = df['species'].replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```

```python
df.head()
```

```python
df.tail()
```

```python
X = df.drop('species', axis=1)
Y = df['species']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

> **2. Create and Train KNN Classification Model :**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_clf.fit(x_train, y_train)
```

> **3. Predict Test Set Results :**

```python
y_pred = knn_clf.predict(x_test)
```

```python
pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
```

> **4. Model Evaluation Performance :**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
mat = confusion_matrix(y_test, y_pred)
print(mat)
```

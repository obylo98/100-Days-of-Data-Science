```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Function to Visualize Cluster Distribution**

```python
# Function to Visualize the Clusters

def visualize_clusters(dataset, c_num, colors, name):
    #visulaizing the clusters  
    for i in range(0, c_num):
        plt.scatter(dataset['x'][dataset['color'] == i], 
                dataset['y'][dataset['color'] == i], s = 100, 
                c = colors[i], label = f'Cluster {i+1}')   

    plt.title(f'Clusters of {name} Dataset ({c_num} colors)')  
    plt.xlabel('X')  
    plt.ylabel('Y')  
    plt.legend()  
    plt.show() 
```

# **Dataset 1 : basic1**

```python
basic1 = pd.read_csv('/kaggle/input/clustering-exercises/basic1.csv')
```

```python
basic1.head()
```

```python
basic1['color'].unique()
```

```python
visualize_clusters(basic1, 4, ['red','blue','cyan','magenta'], 'basic1')
```
